```python
pivot_df = df.groupby(["Keyword Notes", "Vibe"])["Vibe"].size().unstack(fill_value = 0)
pivot_df
```

```python

pivot_df.insert(loc = 3 , column = "Grand Total" , value = pivot_df.sum(axis=1))
pivot_df
```

```python
df2 = pivot_df[(pivot_df[1] > 0) & (pivot_df[2] > 0) & (pivot_df[3] > 0)]
```

```python
df2.insert(loc = 1 , column = "(blank)" , value = "")
df2.head()
```

```python


divide_value = df2["Grand Total"].iloc[0]
# Remove rows where Keyword Notes is blank or 0


x = np.round(df2[1].div(divide_value).mul(100).fillna(0))
y = np.round(df2[2].div(divide_value).mul(100).fillna(0))
z = np.round(df2[3].div(divide_value).mul(100).fillna(0))

df2["Positive"] = x.apply(lambda val: f"{int(val)}%")
df2["Mixed"] = z.apply(lambda val: f"{int(val)}%")
df2["Negative"] = y.apply(lambda val: f"{int(val)}%")
```

```python
df2.head()
```

```python
df2.insert(loc=5, column="Neutral", value="0%")
```

```python
df2.head()
```

```python
df2["Total"] = (
    df2[["Positive", "Negative", "Mixed"]]
    .apply(lambda row: sum(int(val.replace("%", "")) for val in row), axis=1)
    .astype(str) + "%"
)
```

```python
df2.rename({"Total" : "Grand Total"} , inplace = True)
```
