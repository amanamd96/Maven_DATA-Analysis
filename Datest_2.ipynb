```python
df2 = df.groupby(["Age" , "Keyword"])["Keyword"].count().unstack(fill_value = 0)
```

```python
df2.columns.name = None
```

```python
df2
```

```python
df2.rename({"Total" : "Grand Total"} , axis = 1 , inplace = True)
```

```python
df2
```

```python
df2.loc[""] = df2.sum()
df2
```

```python
intender_value = df2.loc["" , "Intender"]
Owner_value = df2.loc["" , "Owner"]
Rejector_value = df2.loc["" , "Rejector"]

x = np.round(df2["Intender"].div(intender_value).mul(100).fillna(0))
y = np.round(df2["Owner"].div(Owner_value).mul(100).fillna(0))
z = np.round(df2["Rejector"].div(Rejector_value).mul(100).fillna(0))

df2["Intenders"] = x.apply(lambda val : f"{int(val)}%")
df2["Owners"] = y.apply(lambda val : f"{int(val)}%")
df2["Rejectors"] = z.apply(lambda val : f"{int(val)}%")

df2
```

```python
total_value = df2.loc["" , "Grand Total"]

df2["Total"] = np.round(df2["Grand Total"].div(total_value).mul(100).fillna(0)).astype(str) + "%"
```

```python
df2.rename({"Total" : "Grand Total"} , axis = 1 , inplace = True)
```
